This jupyter notebook handles Problem 20: Predict whether two people will want to see each other again after a speed-dating first date this project will be a supervised binary classification system based on 123 features X 8400 instances.

```python
# CELL 1: IMPORT LIBRARIES AND LOAD DATA
import pandas as pd
from scipy.io import arff
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
```

```python
arff_file_path = "speeddating.arff"
data_arff, meta = arff.loadarff(arff_file_path)

# Convert to pandas DataFrame
data = pd.DataFrame(data_arff)


# Decode bytes objects to strings for all object columns
for col in data.select_dtypes([object]).columns:
    data[col] = data[col].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)

# Save as a cleaned CSV
csv_file_path = "speeddating_cleaned.csv"
data.to_csv(csv_file_path, index=False)
```

```python
# Load CSV
data = pd.read_csv(csv_file_path)
print("Dataset Overview:")
print(data.head())
data.info()
```

```python
# CELL 2: INITIAL EXPLORATION AND CLEANING
# Clean column names
data.columns = data.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")

# We drop problematic features such as decision and decision_o as these are directly correlated to match and result in a 100% system
columns_to_drop = ['decision', 'decision_o', 'wave', 'has_null', 'race', 'race_o']
data = data.drop(columns=columns_to_drop, errors='ignore')


#Dropping every derived feature since they are derived from the raw numerical data and they have a poorer correlation to match than the raw numbers
# This is done since i decide to strive for a high precision since class one errors in the case of potential matches seems most damaging
data = data.drop(columns=[col for col in data.columns if col.startswith("d_")], errors='ignore')

# Distribution of target variable
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='match')
plt.title("Distribution of Target Variable (Match)")
plt.xlabel("Match (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

# Select only numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
correlation_matrix = numeric_data.corr()

# Correlation analysis to identify highly correlated features
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap="coolwarm", center=0, annot=False)
plt.title("Feature Correlation Matrix")
plt.show()

# Identify correlation with target
correlation_with_target = numeric_data.corr()['match'].sort_values(ascending=False)
top_positive_corr = correlation_with_target.head(11)
top_negative_corr = correlation_with_target.tail(10)

# Visualizing correlations with the target variable
plt.figure(figsize=(8, 6))
top_positive_corr[1:].plot(kind='bar', color='green', title='Top Positive Correlations with Match')
plt.ylabel('Correlation Coefficient')
plt.show()

plt.figure(figsize=(8, 6))
top_negative_corr.plot(kind='bar', color='red', title='Top Negative Correlations with Match')
plt.ylabel('Correlation Coefficient')
plt.show()

top_positive_corr, top_negative_corr

```

```python
top_features = top_positive_corr.index.tolist() + top_negative_corr.index.tolist()

# Heatmap for top correlated features
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data[top_features].corr(), annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title('Heatmap of Top Correlated Features')
plt.show()
```

```python
# Separate features (X) and target (y)
X = data.drop(columns=['match'])
y = data['match']

# Stratified split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
#PreProcessing
# Corrected function for class-based median imputation
def class_based_imputation(X, y, features):
    X = X.copy()
    for feature in features:
        if X[feature].isnull().any():
            # Compute median per class
            medians = X.groupby(y)[feature].median()
            # Apply median imputation class-wise
            for cls in medians.index:
                X.loc[(y == cls) & (X[feature].isnull()), feature] = medians[cls]
    return X

# Features with missing values
features_with_missing = [col for col in X_train.columns if X_train[col].isnull().any()]

# Apply class-based imputation
X_train = class_based_imputation(X_train, y_train, features_with_missing)
X_test = class_based_imputation(X_test, y_test, features_with_missing)

# Verify no missing values
missing_X_train = X_train.isnull().sum().sum()
missing_X_test = X_test.isnull().sum().sum()

print(f"Missing values in X_train: {missing_X_train}")
print(f"Missing values in X_test: {missing_X_test}")
```

```python
# Initial Preprocessing Steps
data = pd.read_csv(csv_file_path)
# 1. Remove unnecessary and derived features as decided earlier
columns_to_drop = ['decision', 'decision_o', 'wave', 'has_null', 'race', 'race_o']
data = data.drop(columns=columns_to_drop, errors='ignore')

# Drop derived features (columns starting with "d_")
derived_features = [col for col in data.columns if col.startswith("d_")]
data = data.drop(columns=derived_features, errors='ignore')

# 2. Separate features (X) and target (y)
X = data.drop(columns=['match'])
y = data['match']

# 3. Split data into training and test sets with stratification
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 4. Drop rows with missing values in both training and test sets
X_train = X_train.dropna()
y_train = y_train.loc[X_train.index]
X_test = X_test.dropna()
y_test = y_test.loc[X_test.index]

# 5. One-hot encode categorical variables
X_train_encoded = pd.get_dummies(X_train, drop_first=True)
X_test_encoded = pd.get_dummies(X_test, drop_first=True)

# Align test set columns with training set
X_test_encoded = X_test_encoded.reindex(columns=X_train_encoded.columns, fill_value=0)

# Dataset ready for scaling and modeling
X_train_encoded.info(), X_test_encoded.info()
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_encoded)
X_test_scaled = scaler.transform(X_test_encoded)

# Scaling completed; dataset is now ready for modeling
X_train_scaled.shape, X_test_scaled.shape
```

```python
rf_baseline = RandomForestClassifier(random_state=42, n_estimators=100)
rf_baseline.fit(X_train_scaled, y_train)

# Extract feature importances
feature_importances = pd.DataFrame({
    'Feature': X_train_encoded.columns,
    'Importance': rf_baseline.feature_importances_
}).sort_values(by='Importance', ascending=False)

# Select top features with importance > 0.01
top_features = feature_importances[feature_importances['Importance'] > 0.01]['Feature']

# Filter datasets to include only top features
X_train_top = X_train_encoded[top_features]
X_test_top = X_test_encoded[top_features]

# Scale the filtered datasets
X_train_top_scaled = scaler.fit_transform(X_train_top)
X_test_top_scaled = scaler.transform(X_test_top)

# Train Logistic Regression and Random Forest models using top features
# Logistic Regression
log_reg_top = LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced')
log_reg_top.fit(X_train_top_scaled, y_train)
log_reg_preds_top = log_reg_top.predict(X_test_top_scaled)

# Random Forest with top features
rf_clf_top = RandomForestClassifier(random_state=42, n_estimators=100, class_weight='balanced')
rf_clf_top.fit(X_train_top, y_train)
rf_clf_preds_top = rf_clf_top.predict(X_test_top)
from sklearn.metrics import classification_report
# Generate classification reports
log_reg_report_top = classification_report(y_test, log_reg_preds_top, target_names=["No Match", "Match"])
rf_clf_report_top = classification_report(y_test, rf_clf_preds_top, target_names=["No Match", "Match"])

print(feature_importances.head())

log_reg_report_top, rf_clf_report_top
```

```python
# Apply SMOTE to balance the classes in the training set
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)

from sklearn.ensemble import GradientBoostingClassifier
# Train a Gradient Boosting Classifier on the balanced dataset
gb_clf_balanced = GradientBoostingClassifier(random_state=42, n_estimators=100, learning_rate=0.1)
gb_clf_balanced.fit(X_train_balanced, y_train_balanced)

# Make predictions on the test set
gb_clf_balanced_preds = gb_clf_balanced.predict(X_test_scaled)

# Generate classification report for the balanced Gradient Boosting model
gb_clf_balanced_report = classification_report(y_test, gb_clf_balanced_preds, target_names=["No Match", "Match"])

gb_clf_balanced_report
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier

# Define hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 500],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(
    GradientBoostingClassifier(random_state=42),
    param_grid,
    cv=3,  # Cross-validation folds
    scoring='f1_macro',
    n_jobs=-1  # Use all available CPU cores
)

# Fit the model
grid_search.fit(X_train_balanced, y_train_balanced)

# Best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best F1 Macro Score:", grid_search.best_score_)
```

```python
# Initialize Gradient Boosting with the best parameters
final_gb_clf = GradientBoostingClassifier(
    learning_rate=0.1,
    max_depth=10,
    min_samples_leaf=2,
    min_samples_split=5,
    n_estimators=500,
    random_state=42
)

# Train the model on the training set
final_gb_clf.fit(X_train_balanced, y_train_balanced)

# Make predictions on the test set
final_gb_preds = final_gb_clf.predict(X_test_scaled)

# Evaluate the model
from sklearn.metrics import classification_report
final_gb_report = classification_report(y_test, final_gb_preds, target_names=["No Match", "Match"])
print(final_gb_report)
```

```python
from collections import Counter
class_counts = Counter(y_train_balanced)
majority_class_size = class_counts[0]  # Count of 'No Match'
minority_class_size = class_counts[1]  # Count of 'Match'

# Calculate scale_pos_weight
scale_pos_weight = majority_class_size / minority_class_size

xgb_clf = XGBClassifier(
    scale_pos_weight=scale_pos_weight,
    learning_rate=0.1,
    max_depth=10,
    n_estimators=500,
    random_state=42
)
xgb_clf.fit(X_train_balanced, y_train_balanced)
xgb_preds = xgb_clf.predict(X_test_scaled)

xgb_report = classification_report(y_test, xgb_preds, target_names=["No Match", "Match"])
print(xgb_report)
```

```python
from xgboost import plot_importance
plot_importance(xgb_clf, max_num_features=10)  # Visualize top 10 features
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, xgb_preds)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["No Match", "Match"], yticklabels=["No Match", "Match"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# CELL 5: FEATURE SELECTION WITH LOGISTIC REGRESSION
# Train Logistic Regression to identify top features
logistic_model = LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42)
logistic_model.fit(X_train, y_train)

# Extract feature importance
feature_importance = pd.Series(
    np.abs(logistic_model.coef_[0]), index=X_train.columns
).sort_values(ascending=False)

# Visualize top 20 features
plt.figure(figsize=(10, 6))
feature_importance.head(20).plot(kind='bar', color='blue')
plt.title("Top 20 Most Important Features (Logistic Regression)")
plt.xlabel("Features")
plt.ylabel("Coefficient Magnitude")
plt.show()

# Select top 50 features
top_features = feature_importance.head(20).index.tolist()

# Reduce the dataset to top features
X_train_reduced = X_train[top_features]
X_test_reduced = X_test[top_features]
```

```python
# CELL 6: BASELINE MODEL - LOGISTIC REGRESSION
# Evaluate Logistic Regression
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    return {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred),
        "ROC-AUC": roc_auc_score(y_test, y_proba),
    }

logistic_metrics = evaluate_model(logistic_model, X_train_reduced, y_train, X_test_reduced, y_test)
print("Logistic Regression Metrics:", logistic_metrics)
```

```python
# CELL 7: RANDOM FOREST
# Train and evaluate Random Forest
random_forest_model = RandomForestClassifier(random_state=42, n_estimators=100)
random_forest_metrics = evaluate_model(random_forest_model, X_train_reduced, y_train, X_test_reduced, y_test)
print("Random Forest Metrics:", random_forest_metrics)
```

```python
# CELL 8: XGBOOST
# Train and evaluate XGBoost
xgb_model = XGBClassifier(
    use_label_encoder=False,
    eval_metric="logloss",
    n_estimators=100,  # Reduce this for quicker runs
    max_depth=6,       # Limit depth for faster training
    random_state=42
)
xgb_model.fit(X_train_reduced, y_train)
xgb_metrics = evaluate_model(xgb_model, X_train_reduced, y_train, X_test_reduced, y_test)
print("XGBoost Metrics:", xgb_metrics)
```
